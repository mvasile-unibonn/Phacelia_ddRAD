# Tetrad Analysis for ddRAD assembly of the ingroup
A species tree inference tool based on the SVDQuartets algorithm resolves quartet trees from SNPs for all sets of quartets in a larger tree, and then joins the quartets together into a supertree.

### we will run a multiple tetRAD analysis in a loop for all different pops of cluster threshold 85, then 90, then 95

```python
## tetrad for clust. threshold 85
## define the path to your sequence data in HDF5 format
## The input data file should be the .snps.hdf5 
## read the *.snps.hdf5 files as values and store those path in a dictionary with assembly names as keys
dict = {
    "pop15": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops15_clust85.snps.hdf5",
    "pop20": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops20_clust85.snps.hdf5",
    "pop25": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops25_clust85.snps.hdf5",
    "pop30": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops30_clust85.snps.hdf5",
    "pop35": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops35_clust85.snps.hdf5",
    "pop40": "/home/marianna/Documents/Phacelia/hdf5_files_202209/pops40_clust85.snps.hdf5"
}
```

```python
## Iterate through the dictionary and run a tetRAD anlysis for each assembly

for key, value in dict.items():
    tet = ipa.tetrad(
        name = "Phac_tet_clust85_" + str(key),
        data = value,
        workdir = "./Phac_Analyses/tetrad/tet_clust85",
        nquartets = 1e6,
        nboots = 200)
    ## run 
    tet.run(auto = True, force = True)
```

```python
## Plot all six clust85 tetRAD coalescent trees together
## Load trees
tet15 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop15.tree.cons")
tet20 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop20.tree.cons")
tet25 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop25.tree.cons")
tet30 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop30.tree.cons")
tet35 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop35.tree.cons")
tet40 = toytree.tree("/home/marianna/Documents/Phacelia/Phac_Analyses/tetrad/tet_clust85/Phac_tet_clust85_pop40.tree.cons")

tet15 = tet15.root(wildcard = "Phacelia_artemisioides_W6021_out")
tet20 = tet20.root(wildcard = "Phacelia_artemisioides_W6021_out")
tet25 = tet25.root(wildcard = "Phacelia_artemisioides_W6021_out")
tet30 = tet30.root(wildcard = "Phacelia_artemisioides_W6021_out")
tet35 = tet35.root(wildcard = "Phacelia_artemisioides_W6021_out")
tet40 = tet40.root(wildcard = "Phacelia_artemisioides_W6021_out")

## set dimensions of the canvas
canvas = toyplot.Canvas(width = 2000, height = 2000)

## dissect canvas into multiple cartesian areas (x1, x2, y1, y2)
ax0 = canvas.cartesian(bounds=('2%',  '30%', '5%',  '47.5%'))
ax1 = canvas.cartesian(bounds=('33%', '63%', '5%',  '47.5%'))
ax2 = canvas.cartesian(bounds=('66%', '96%', '5%',  '47.5%'))
ax3 = canvas.cartesian(bounds=('2%',  '30%', '50%', '97.5%'))
ax4 = canvas.cartesian(bounds=('33%', '63%', '50%', '97.5%'))
ax5 = canvas.cartesian(bounds=('66%', '96%', '50%', '97.5%'))

## define style ones and use it for every tree
style = {
    "tip_labels_align": True,
    "tip_labels_style": {"font-size": "11px"},
    "node_labels_style":{"font-size": "12px",
                        "baseline-shift": "7px",
                        "-toyplot-anchor-shift": "-13px"},
}
tet15.ladderize(1).draw(
    axes = ax0,
    **style,
    node_sizes = 0,
    node_labels = 'support');

tet20.ladderize(1).draw(
    axes = ax1,
    **style,
    node_sizes = 0,
    node_labels = 'support');

tet25.ladderize(1).draw(
    axes = ax2,
    **style,
    node_sizes = 0,
    node_labels = 'support');

tet30.ladderize(1).draw(
    axes = ax3,
    **style,
    node_sizes = 0,
    node_labels = 'support');

tet35.ladderize(1).draw(
    axes = ax4,
    **style,
    node_sizes = 0,
    node_labels = 'support');

tet40.ladderize(1).draw(
    axes = ax5,
    **style,
    node_sizes = 0,
    node_labels = 'support');

## hide the axes (e.g, ticks and splines)
ax0.show = False; ax1.show = False; ax2.show = False;
ax3.show = False; ax4.show = False; ax5.show = False;

## add names for the single trees
canvas.text(1000, 50, 'tetRAD/SVDQuartet - Clustering threshold 85 %', style = {"font-size": "24px"})
canvas.text(150, 125, '85 % missing data', style={"font-size": "18px"})
canvas.text(800, 125, '80 % missing data', style={"font-size": "18px"})
canvas.text(1450, 125, '75 % missing data', style={"font-size": "18px"})
canvas.text(150, 1025, '70 % missing data', style={"font-size": "18px"})
canvas.text(800, 1025, '65 % missing data', style={"font-size": "18px"})
canvas.text(1450, 1025, '60 % missing data', style={"font-size": "18px"});
```
